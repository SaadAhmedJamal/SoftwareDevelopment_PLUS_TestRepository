```python
import folium
from folium import GeoJson
```

```python
from folium import GeoJson
geo=r"Grid.geojson"
file = open(geo, encoding="utf8")
text = file.read()
m = folium.Map(width="%100",weight="%100", tiles='Stamen Toner')
GeoJson(text).add_to(m)
```

```python
marker_location = [33.824, 72.045]
tooltip = "NUST Islamabad"
popup = "<b>Cloud!</b>"

marker_location1 = [47.499, 13.04]
tooltip1 = "University of Salzburg"
popup = "<b>Cloud!</b>"

marker_location2 = [47.409, 2.45]
tooltip2 = "University of South Brittany"
popup = "<b>Cloud!</b>"
folium.LayerControl().add_to(m)

folium.Marker(marker_location, popup=popup, tooltip=tooltip, draggable=True, icon=folium.Icon(color="red", icon="cloud")).add_to(m)
folium.Marker(marker_location1, popup=popup, tooltip=tooltip1, draggable=True, icon=folium.Icon(color="red", icon="cloud")).add_to(m)
folium.Marker(marker_location2, popup=popup, tooltip=tooltip2, draggable=True, icon=folium.Icon(color="red", icon="cloud")).add_to(m)
```
